```python
target=pd.read_csv('updatedtest1.csv')
```

```python
target
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression(max_iter=10000)
```

```python
model1.fit(newtrain.loc[:,'xyz_concern':'age_group_weight'], newtrain.loc[:,'xyz_vaccine'])
```

```python
pred_prob_xyz = model1.predict_proba(target.loc[:,'xyz_concern':'age_group_weight'])
```

```python
pred_prob_xyz[:,1]
```

```python
pred_prob_xyz_train = model1.predict_proba(newtrain.loc[:,'xyz_concern':'age_group_weight'])
```

```python
from sklearn.metrics import roc_auc_score
auc_score1_train = roc_auc_score(newtrain.loc[:,'xyz_vaccine'], pred_prob_xyz_train[:,1])
print(auc_score1_train)
```

```python
model1.fit(newtrain.loc[:,'xyz_concern':'age_group_weight'], newtrain.loc[:,'seasonal_vaccine'])
```

```python
pred_prob_seasonal = model1.predict_proba(target.loc[:,'xyz_concern':'age_group_weight'])
```

```python
pred_prob_seasonal[:,1]
```

```python
pred_prob_seasonal_train = model1.predict_proba(newtrain.loc[:,'xyz_concern':'age_group_weight'])
```

```python
auc_score2_train = roc_auc_score(newtrain.loc[:,'seasonal_vaccine'], pred_prob_seasonal_train[:,1])
print(auc_score2_train)
```
